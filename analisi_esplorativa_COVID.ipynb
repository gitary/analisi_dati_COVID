# Metriche dashboard desktop e mobile

```python
!ls metriche/dpc-covid19-ita-metriche-dashboard-desktop.csv
```

```python
df_metriche_desktop = pd.read_csv('metriche/dpc-covid19-ita-metriche-dashboard-desktop.csv', encoding='latin-1')
df_metriche_mobile = pd.read_csv('metriche/dpc-covid19-ita-metriche-dashboard-mobile.csv', encoding='latin-1') 
```

```python
df_metriche_desktop.info()
```

```python
df_metriche_desktop
```

```python
df_metriche_desktop.set_index('data')
df_metriche_mobile.set_index('data')
```

```python
fig, ax = plt.subplots(figsize=(14,8))
df_metriche_desktop.plot(x = 'data', y = 'totale_visitatori_giorno', label ='desktop', ax=ax)
df_metriche_mobile.plot(x = 'data', y = 'totale_visitatori_giorno', label ='mobile', ax = ax)
plt.show()
```

```python
df_metriche_desktop.shape, df_metriche_mobile.shape
```

```python
df_metriche_desktop.columns
```

```python
col_mob = []
for ele in df_metriche_mobile.columns: 
    col_mob.append(ele+'_m')
col_mob
```

```python
df_sup_mobile = df_metriche_mobile.copy()
df_sup_mobile.columns = col_mob
```

```python
df_metriche = pd.concat([df_metriche_desktop, df_sup_mobile], axis = 1)
df_metriche.loc[:,'data'] = pd.to_datetime(df_metriche['data'], format='%d/%m/%Y')
```

```python
df_metriche
```

```python
pandas_bokeh.output_notebook()

#y = ['totale_visitatori_giorno', 'totale_visitatori_giorno_m']
g = df_metriche.plot_bokeh(x='data', y = ['totale_visitatori_giorno', 'totale_visitatori_giorno_m',\
                                'visitatori_picco_ore_18', 'visitatori_picco_ore_18'],\
                           vertical_xlabel = True, title="Visitatori dashboard", figsize=(950, 500))
#g = df_metriche_mobile.plot_bokeh(x = 'data', y = 'totale_visitatori_giorno', vertical_xlabel = True, title="Visitatori dashboard", figsize=(950, 500))
```

```python
g = df_metriche.plot_bokeh(x='data',y = ['totale_visitatori_giorno', 'totale_visitatori_giorno_m'],\
                           vertical_xlabel = True, title="Visitatori dashboard", figsize=(950, 500))
```

```python
g = df_metriche.plot_bokeh(x='data',y = ['visitatori_picco_ore_18', 'visitatori_picco_ore_18_m'],\
                           vertical_xlabel = True, title="Visitatori dashboard", figsize=(950, 500), logy=True)
```

```python
data_resample_M = df_metriche.set_index('data').resample(rule='M').mean()
```

```python
data_resample_W = df_metriche.set_index('data').resample(rule='W').mean()
```
